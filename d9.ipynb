```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset
```

```python
# Load the Titanic dataset
data = load_dataset('titanic')
```

```python
# Task 1: Use Seaborn library to visualize patterns in the data
# Let's create a pairplot to visualize relationships between numerical variables
sns.pairplot(data, hue='survived')
plt.title('Pairplot of Titanic Dataset')
plt.show()
```

```python
# Task 1: Use Seaborn library to visualize patterns in the data
# Let's create a pairplot to visualize relationships between numerical variables
sns.pairplot(data, hue='sex')
plt.show()
```

```python
# Task 2: Plot a histogram to visualize the distribution of ticket prices (fare)
plt.figure(figsize=(10, 6))
sns.histplot(data['fare'], bins=20, kde=True, color='blue')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()


# For example, setting bins=20 in a histogram will divide the data range into 20 intervals, 
# each representing a fraction of the total range of fare values in this case. 
# Adjusting the bins parameter allows you to explore the data distribution with different levels of granularity.
```
