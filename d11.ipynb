```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
```

```python
# 1. List down the features and their types
# These lines print out the features (columns) of the dataset along with their corresponding data types.
print("Features and their types:")
print(df.dtypes)
```

```python
# 2. Create a histogram for each feature
# These lines create histograms for each feature in the dataset and display them using matplotlib. 
# figsize sets the size of the figure, and suptitle adds a centered title above the subplots.
df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

```python
# 3. Create a boxplot for each feature
# These lines create boxplots for each feature in the dataset using seaborn. 
# figsize sets the size of the figure, boxplot creates the boxplots, and title sets the title of the plot.


plt.figure(figsize=(10, 8))
sns.boxplot(data=df)
plt.title("Boxplots of Iris Dataset Features")
plt.show()
```

```python
# 4. Compare distributions and identify outliers
# For outlier detection, let's use boxplots
# These lines specifically create boxplots for the numerical features 
# (sepal length, sepal width, petal length, and petal width) to visually identify potential outliers. 
# This helps in comparing the distributions of these features and identifying any 
# data points that fall significantly outside the whiskers of the boxplots, which could indicate outliers.


plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']])
plt.title("Boxplots of Iris Dataset Features")
plt.show()
```
