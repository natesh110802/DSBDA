```python
# Convert 'Data_value' column to numeric type
df['Data_value'] = pd.to_numeric(df['Data_value'], errors='coerce')
```

```python
# Using one-hot encoding for 'Anzsic_description' column
df_encoded = pd.get_dummies(df, columns=['Anzsic_description'])
```
